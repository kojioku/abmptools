# abmptools.hbond — IMC amorphous sample

非晶質インドメタシン (IMC) の MD スナップショットに対する H-bond 検出 + 分類 + UDF 色付け。

操作手順:
1. 下のセルを実行 → パネル表示
2. パネル上で検出基準 (Luzar-Chandler / strict / custom) を選択
3. `Run` ボタンクリック
4. 統計表 + count plot が表示される
5. 出力された `*_colored.bdf` を gourmet で開いて 3 色可視化を確認

```python
from abmptools.hbond import open_panel

# Input trajectory is bundled at sample/hbond/imc_amorphous/input/.
# When this notebook is opened from that directory, the relative path resolves.
import os
BDF = os.path.abspath('input/IMC_result450.0_out_rec900.bdf')
panel = open_panel(BDF)
```

## Python API 直接利用 (任意)

```python
from abmptools.hbond import Analyzer, AnalyzerConfig

cfg = AnalyzerConfig(
    bdf_path=BDF,
    out_prefix='./output/imc_hbond_api',
    criteria_mode='luzar-chandler',
    base_mol_name='IMC',
)
a = Analyzer(cfg)
a.load()
results = a.run()
paths = a.write_outputs()
print('Output files:', paths)
```
