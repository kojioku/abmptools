# abmptools.hbond — Jupyter UI sample (IMC default)

abmptools.hbond の GUI 解析パネルを起動するサンプル notebook。
default では bundle 済みの非晶質インドメタシン (IMC) を読み込むが、
**環境変数 `HBOND_BDF` を渡せば任意の BDF / UDF に切替**できる:

```
# PowerShell:  $env:HBOND_BDF="C:\path\to\my.bdf"; jupyter notebook run_notebook.ipynb
# cmd:         set HBOND_BDF=C:\path\to\my.bdf && jupyter notebook run_notebook.ipynb
# bash:        HBOND_BDF=/path/to/my.bdf jupyter notebook run_notebook.ipynb
```

操作手順:
1. 下のセルを実行 → パネル表示
2. パネル上で:
   - Mode dropdown (`imc` / `generic`) で分類モード切替
   - mol prefix を別系名 (例: `PVA`) に編集
   - Donor / Acceptor checkboxes で官能基組合せを選択
   - 検出基準 (Luzar-Chandler / strict / custom)
3. `Run` ボタンクリック
4. 統計表 + count plot が表示される
5. 出力された `*_colored.bdf` を OCTA gourmet で開いて 3 色可視化を確認


````python
from abmptools.hbond import open_panel
import os

# Default: bundled IMC sample. Override with HBOND_BDF env var
# (see top markdown for PowerShell / cmd / bash examples).
_DEFAULT_BDF = os.path.abspath('input/IMC_result450.0_out_rec900.bdf')
BDF = os.environ.get('HBOND_BDF', _DEFAULT_BDF)
print(f'Input BDF: {BDF}')
panel = open_panel(BDF)
````

## Python API 直接利用 (任意)

````python
from abmptools.hbond import Analyzer, AnalyzerConfig

cfg = AnalyzerConfig(
    bdf_path=BDF,
    out_prefix='./output/imc_hbond_api',
    criteria_mode='luzar-chandler',
    base_mol_name='IMC',
)
a = Analyzer(cfg)
a.load()
results = a.run()
paths = a.write_outputs()
print('Output files:', paths)
````
